```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Use the correct method to set the style
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# Continue with the rest of your code
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) 

df=pd.read_csv(r"C:\Users\DELL\Downloads\movies.csv")
df.head()
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head()
```

```python
# Extract the 4-digit year from 'released' and ensure it's a string type
extracted_years = df['released'].str.extract(pat='([0-9]{4})').astype(str)

# Print the extracted values to check correctness
print("Extracted Years:")
print(extracted_years.head())

# Assign the extracted years, converting them to integer after handling NaNs
#df.loc[:, 'yearcorrect'] = extracted_years[0].fillna(0).astype(int)
df['yearcorrect'] = extracted_years[0].fillna(0).astype(int)
# Print the final DataFrame
print("Final DataFrame:")
print(df[['released', 'yearcorrect']].head())
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.drop_duplicates().head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budger vs Gross earnings')
plt.xlabel('Gross earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
sns.regplot(x="budget", y="gross", data=df,scatter_kws={"color":"red"},line_kws={"color":"purple"})
```

```python
# Select only the numeric columns
numeric_df = df.select_dtypes(include=[float, int])
correlation_matrix = numeric_df.corr(method='pearson')

# Display the result
print(correlation_matrix)
# pearson kendall spearman
```
